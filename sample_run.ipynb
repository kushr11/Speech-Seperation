```python
# Dataset: TIMIT
#* Dataset dir: /home/xinying/Speaker_L/TIMIT_Dataset/data/augmented 
# 0-3ppls, resized to 10 seconds, with overlay or concateate, without data augmented
#* Dataset dir: /home/xinying/Speaker_L/TIMIT_Dataset/data/audio_augmented4  
# 0-3ppls, resized to 10 seconds, with overlay or concateate, with data augmented
#* Dataset dir: /home/xinying/Speaker_L/TIMIT_Dataset/data/raw_augmented
# 0-3ppls, not resized, with overlay or concateate
#* Dataset dir: /home/xinying/Speaker_L/TIMIT_Dataset/data/separa_cancat23
# Seperated noise and speech data by the model, with origin concatenated 2-3 ppls data
audio_path="/home/xinying/Speaker_L/TIMIT_Dataset/data/augmented/2-MTCS0_MPAM0=overlay_SX262.WAV" 
```

```python
import soundfile
from IPython.display import display, Audio
mixwav, sr = soundfile.read(audio_path)
display(Audio(mixwav, rate=sr))
```

```python
from speechbrain.inference.separation import SepformerSeparation as separator
model = separator.from_hparams(source="speechbrain/sepformer-libri3mix", savedir='pretrained_models/sepformer-libri3mix')
est_sources = model.separate_file(path=audio_path)
```

```python
Audio(est_sources[:, :, 1].detach().cpu().squeeze(), rate=8000) #0-2
```

```python
soundfile.write('audio_trimmed.wav', est_sources[:, :, 1].detach().cpu().squeeze(), 8000) #save
```
